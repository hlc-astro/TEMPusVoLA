We would like to check how much powder, of a given porosity, is required to induce a measurable pressure difference when gas flows through it. We make initial estimates based upon simple navier-stokes equation with linear drag force,  that is mediated by a permeability coefficient $\chi$. We impose conservation of mass flow rate, Q. The subscript u refers to 'upper' pressure and l to 'lower' pressure. 

$\rho_{u}a_{u}v_{u}=\rho_{l}a_{l}v_{l}=Q\\
[eq. 1]  \rho_{u}a_{u}v_{u}=\rho(h)a(h)v(h)\\
[eq. 2]   \frac{dP}{dh}=-\frac{\mu}{\chi}v(h)\\
v(h) =\frac{v_{l}P_{l}a_{l}}{P(h) a(h)}\\
\int_{P_{u}}^{P_{l}}{P(h)}dP=-\frac{\mu}{\chi} v_{l}P_{l}\int_{h(P=P_{u})}^{h(P=P_{l})}{}dh\\
\frac{P^{2}}{2}|^{P_{l}}_{P_{u}} +C_{1}=-\frac{\mu v_{l}P_{l}}{\chi}h+C_{2}\\
\text{At boundary, } C_{1}=C_{2}\\
[eq. 3]\frac{P^{2}_{l}}{2}-\frac{P^{2}_{u}}{2}=-\frac{\mu v_{l}P_{l}}{\chi}dz\\
\frac{(P^{2}_{l}+P^{2}_{u})(P^{2}_{l}-P^{2}_{u})}{2}=-\frac{\mu v_{l}P_{l}}{\chi}dz\\
\chi=\frac{\mu v_{l}P_{l}\delta z}{|\delta P| \frac{P_{2}+P_{1}}{2}}= \frac{\mu P_{l}\delta z}{|\delta P| \frac{P_{2}+P_{1}}{2}} \frac{Q}{a_{l}\rho_{l}}\\
\rho_{l}=\frac{P_{l}M}{RT}\\
\chi= \frac{\mu\delta z}{|\delta P| \frac{P_{2}+P_{1}}{2}} \frac{Q}{a_{l}}\frac{RT}{M}$

Here R,T,and M are the gas constant, temperature, and mean molecular mass, respectively. $\delta z$ is the height of the bed. By measuring P$_{l}$ and P$_{u}$, and with known height and mass flow rate (or linear flow rate, v), $\chi$ is easily extracted. The units of this quantity are m$^{2}$ and it refers to the average cross-sectional pore opening space (i.e. the dimension of the channels through which the flow can pass). For spherical particles and a flow with low Reynold's number, the equivalent quantity can be derived using the Carman-Kozenzy correlation, $\xi_{spheres}=K_{coz} d_{p}^{2}(1-c)^{3}/c^{2}$. c is the porosity, and relateds to f, the filling factor, c+f=1. For a random-packed bed of spheres of equal size, f is always 0.64 and K$_{coz}$ is 2/90.  

We can show the equivalence of these two coefficients and simultaneously test the pressure drop we should expect with height, by assuming that eq. 3 holds not only at P_{u} and P_{l}, but by all heights in the bed. 

The pressure as a function of height is then given by: 

$P(h)=\sqrt{P_{l}^{2}-\left(\frac{2\mu v_{l}P_{l}}{\chi}\right)h}$

We imagine using 0.5 mm steel spheres as calibration material, and plot the pressure with height in the bed for this particle size ($\chi_{sphere}$), and hunt for the equivalent $\chi$.

We see that for $\delta z =20$cm, we can take the pressure down from 10 mbar to 1 mbar. We have sensitivity in our pressure sensors on the order of 20 percent, and so this difference is easy to detect as would, for example the difference from 5-1 mbar, which is given by a 5 cm height. We note that at these pressures, a .5 mm particle has a Kn of 0.1-1, and so that slip effects start to become important.  

It is common to assume a correction to the permeability of the form $\chi_{\rm Klinkenberg}=\chi\left(1+\frac{b}{\frac{P_{u}-P_{l}}{2}} \right)$. The exact form of b is highly debated, but an example of the relation ship is: $b=0.7\chi^{0.3}$. Therefore, we see that for the klinkenberg correction becomes relevant if the order of the permittivity is very small compared to the order of the pressure. For the case plotted above, the black line is for $\chi=7.5\times10^{-10}$, and the mean pressure is order 500 pa, hence the correction term, b/p, would be around 2500/500=5, i.e. less than an order of magnitude difference. For the same particle size and packing fraction, each decade decrease in gas pressure would then increase the relevance of the correction in proportionate measure.
