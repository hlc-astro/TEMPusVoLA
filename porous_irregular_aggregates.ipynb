We now ask practical questions regarding the terminal velocity, stopping time, and stopping distance, of particles in the parameter space above, when the gravitational acceleration is allowed to change. 

We calculate the terminal velocity using the Stokes drag force, $v_{Stokes}=\frac{1}{18}g\eta d_{p}^2\rho$, and apply a Kn-dependent correction of the form: $C_{Kn}=1.0+Kn(\alpha+\beta e^{-\gamma/Kn})$, so that the terminal velocity depends upon Kn and is $v(Kn)=C_{Kn}v_{Stokes}$. Here $g=9.8$ m/s$^{2}$ is Earth's gravitational acceleration, and $\eta$ is the molecular viscosity of air. $\alpha$, $\beta$, and $\gamma$ are taken from the experimental literature. 

Since we are interested in order of magnitude estimates, we do not account for porosity, and consider three regimes in material density, that of ice (~1000 kg/m$^{2}$; solid lines), rock (~3000 kg/m$^{2}$; dashed lines), and metal (~8000 kg/m$^{2}$;dot-dashed lines). (Note that the photoresist particles have density ~1200 kg/m$^{2}$, and so are close to the density of ice and less than that of rock.) 

We now express quantities in terms of the friction time $T_{f}$, which is the ratio of the particle momentum to its drag force, and determines the scale time over which the particle reaches its termial velocity, $v(Kn)=T_{f}g$. The associated distance scale over which a particle reaches its steady state velocity is the friction length, $L(Kn)=T_{f}^{2}g$. 

```python
Image(filename='Figure_1.png')
```

Evidently, particles in the 10-100 micron range have extremely high terminal velocities (several 10s of m/s) when they are approaching drag in the free-molecular flow regime, according to the above, around 0.1-0.5 mbar. However, since the terminal velocity scales directly with g, as the gravitational force decreases, so does the velocity of the particle. We see that for 1/100 the gravitational acceleration of Earth, particles under free-molecular flow conditions may be moving centimeters per second. Note that the reported "jitter" on the plane is on the order of 10$^{-3}$g, and so we can expect, accordingly, variations in the gravity that could induce variations of mm/s ... however, I suspect this is mainly an issue at the instant when the particle is released into the chamber.
